```python
import spotipy
import pandas as pd
import numpy as np
import json
from collections import defaultdict

artists = pd.read_pickle('helper/artist_df.pkl')
```

```python
spotify = spotipy.Spotify()
spotifydict = {}
i = 0
```

```python
start = i

for i in range(start, len(artists.name)) :
    artist_name = artists.name[i]
    if pd.isnull(artist_name) == False :
        spotifydict[artist_name] = {}
        try :
            results = spotify.search(q='artist:' + artist_name, type='artist')
        except :
            results = []
        if len(results) > 0 :
            if len(results['artists']['items']) != 0 :
                spotifydict[artist_name]['genres'] = results['artists']['items'][0]['genres']
                spotifydict[artist_name]['popularity'] = results['artists']['items'][0]['popularity']
                spotifydict[artist_name]['followers'] = results['artists']['items'][0]['followers']
                print "Found: " + artist_name
            else :
                print "Unknown: " + artist_name
```

```python
import json
with open('helper/spotify.json', 'w') as f:
    json.dump(spotifydict, f)
```
