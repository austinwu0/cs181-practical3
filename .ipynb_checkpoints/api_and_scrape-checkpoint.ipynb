```python
import pandas as pd
import numpy as np
import requests
import time
import re
import json
from bs4 import BeautifulSoup
from collections import defaultdict

artists = pd.read_pickle('helper/artist_df.pkl')
musicbrainDict = {}
i = 0
```

```python
start = i

for i in range(start, len(artists.name)) :
    artist_name = artists.name[i]
    if pd.isnull(artist_name) == False :
        musicbrainDict[artist_name] = {}

        ID = artists.artist_id[i]
        URL = "https://musicbrainz.org/artist/" + ID
        var = requests.get(URL)
        soup = BeautifulSoup(var.text, "html.parser")
        result = soup.find_all("dd")
        for entry in result :
            try :
                info_class = entry.get('class','')[0]
            except IndexError:
                info_class =  "Founded"
            musicbrainDict[artist_name][info_class] = entry.get_text()

        print "Found: " + artist_name
```

```python
import json
with open('helper/musicbrain.json', 'w') as f:
    json.dump(musicbrainDict, f)
```
